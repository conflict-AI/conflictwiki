# TF-IDF Analysis

### import libraries

```python
from wb0configs import configs
from wb0configs.helpers import store_file, load_file

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

config = configs.ConfigBase()
```

## Entity Corpus tf-idf

```python
e_tfidf_features = load_file(config.get_path("entity_embed") / "e_tfidf_features", ftype="pkl")
e_toks = load_file(config.get_path("entity_embed") / "ent_toks", ftype="pkl")
e_id_embed = load_file(config.get_path("task") / "entity_pooled" / "e_id_embed_tfidf", ftype="pkl")
e_id_name = load_file(config.get_path("entity_retrieval") / "entity_id_name", ftype="pkl")
e_tfidf_features
```

### Entities

19127 Mali <br>
5843419 France <br>
34593948 Azawad <br>
1921 Al-Qaeda <br>


```python
feature_array = np.array(list(e_tfidf_features.values()))

for entity_id in [19127, 5843419, 34593948, 1921]:
    tfidf_sorting = np.argsort(e_id_embed[entity_id].detach().numpy()).flatten()[::-1]
    top_n = feature_array[tfidf_sorting][:10]
    print(e_id_name[entity_id], ": ", top_n, "\n")
```
